# CNNs for Image Colorization

```python
import tensorflow as tf
import keras
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import cv2
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D, Activation
from keras.callbacks import ModelCheckpoint, Callback
```

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
```

### 2. (b) From the train and test dataset, extract the class birds. We will focus on this class, which has 6000 members.

```python
train_bird_images = x_train[[i for i in range(len(y_train)) if y_train[i] == [6]]]
test_bird_images = x_test[[i for i in range(len(y_test)) if y_test[i] == [6]]]

bird_images = np.concatenate([train_bird_images, test_bird_images], axis=0)
bird_images.shape
```

### 2. (c) Those 6000 images have 6000 × 32 × 32 pixels. Choose at least 10% of the pixels randomly. It is strongly recommended that you choose a large number or all of the pixels. You will have between P = 614400 and P = 6144000 pixels. Each pixel is an RGB vector with three elements.

```python
P = []
for i in range(bird_images.shape[0]):
    for j in range(bird_images.shape[1]):
        for k in range(bird_images.shape[2]):
            P.append(bird_images[i][j][k])
P = np.array(P)
P.shape
```

### 2. (d) Run k-means clustering on the P vectors using k = 4.

```python
kmeans = KMeans(n_clusters=4, random_state=7)
kmeans.fit(P)

kmeans.cluster_centers_
```

```python
cluster_centers = kmeans.cluster_centers_.astype(int)
cluster_centers
```

```python
Y = np.zeros((6000,32,32,3))
for i in range(bird_images.shape[0]):
    for j in range(bird_images.shape[1]):
        for k in range(bird_images.shape[2]):
            distance = []
            for c in range(len(cluster_centers)):
                distance.append(np.linalg.norm(bird_images[i][j][k] - cluster_centers[c]))
            distance = np.array(distance)
            Y[i][j][k] = cluster_centers[np.argmin(distance)]            
```

### 2. (e) Use any tool (e.g., openCV or scikit-learn) to obtain grayscale 32 × 32 × 1 images from the original 32 × 32 × 3 images. The grayscale images are inputs of your network.

```python
x = np.zeros((6000,32,32,1))

for i in range(x.shape[0]):
    x[i] = np.expand_dims(cv2.cvtColor(bird_images[i], cv2.COLOR_BGR2GRAY), axis=-1)
```

```python
y_encoded = to_categorical(kmeans.labels_)
y_encoded = y_encoded.reshape(6000, 4096)
y_encoded.shape
```

```python
Xtrain, Ytrain = x[:5000], y_encoded[:5000]
Xtest, Ytest = x[5000:], y_encoded[5000:]

Xtrain = Xtrain/255.
Xtest = Xtest/255.
```

```python
model = Sequential()
model.add(Conv2D(64, 5, activation='relu', input_shape=(32,32,1), padding='same'))
model.add(Conv2D(32, 5, activation='relu', padding='same'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(6144, activation='relu'))
model.add(Dropout(0.25))
model.add(Dense(4096, activation='relu'))
model.add(Activation('softmax'))
model.summary()
```

```python
test_loss =[]
class custom_callback(keras.callbacks.Callback):
    def on_epoch_end(self, epoch,logs=None):
        l = self.model.evaluate(Xtest, Ytest)
        test_loss.append(l)
```

```python

callback = [ModelCheckpoint(filepath='best_weights.h5',monitor="val_loss",verbose=1,save_best_only=True, save_freq="epoch"), ]
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['mse'])
history = model.fit(Xtrain, Ytrain, validation_split=0.2, batch_size=32, epochs=10, callbacks=[callback,custom_callback()])
```

```python
pred = model.predict(Xtest[:10])
pred_enc = pred.reshape(10,32,32,4)
pred_enc = np.argmax(pred_enc,axis=-1)
pred_img = kmeans.cluster_centers_[pred_enc]
pred_img = pred_img.astype(int)
pred_img.shape
```

```python
test_loss_ = [i[0] for i in test_loss]
```

```python
training_loss = history.history['loss']
validation_loss = history.history['val_loss']
n=11

plt.figure()
plt.plot(np.arange(1,n),training_loss, label="training_loss")
plt.plot(np.arange(1,n),validation_loss, label="validation_loss")
plt.plot(np.arange(1,n),test_loss_, label="test_loss")
plt.legend()
```
