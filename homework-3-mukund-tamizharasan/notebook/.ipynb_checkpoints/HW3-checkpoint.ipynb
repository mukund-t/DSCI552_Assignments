```python
#defining dataset path and feature names
root_dir = '../data/AReM/'
column_names = ['avg_rss12','var_rss12','avg_rss13','var_rss13','avg_rss23','var_rss23']
dataset_folders = ['bending1','bending2','cycling','lying','sitting','standing','walking']
test_data_ex = ['bending1','bending2']
features = ['mean', 'std', 'min', 'first_quartile', 'median', 'third_quartile','max']
```

#### 1. (b). Train and test data split

```python
#function to split training and test data with the given number of splits
def split_train_test(feature_list,splits=1):
    all_features = []
    for j in range(len(column_names)*splits):
        for i in features:
            all_features.append(i+'_'+str(j+1))
    
    train_instances = []
    test_instances = []
    train_label=[]
    test_label=[]
    for dataset in dataset_folders:
        
        for each_csv in os.listdir(root_dir+dataset):
            if dataset=='bending2' and each_csv=='dataset4.csv':
                df = pd.read_csv(root_dir+dataset+'/'+each_csv,skiprows=5,names=column_names,sep="\s+")
            elif each_csv[-4:]==".csv":
                df = pd.read_csv(root_dir+dataset+'/'+each_csv,skiprows=5,names=column_names)
            
            #findign length of each split
            split_len = int(df.shape[0]/splits)
            
            train_segs=[]
            test_segs=[]
            
            #processing data in each split and assigning to train and test
            for j in range(splits):
                df_des = df[j*split_len : min((j+1)*split_len,df.shape[0])]
                
                if (dataset in test_data_ex and (each_csv in ['dataset1.csv','dataset2.csv'])) or (dataset not in test_data_ex and (each_csv in ['dataset1.csv','dataset2.csv','dataset3.csv'])):
                    df_des = df_des.describe().drop('count').T
                    test_segs.append(df_des.values.flatten())
                elif each_csv[-4:]==".csv":
                    df_des = df_des.describe().drop('count').T
                    train_segs.append(df_des.values.flatten())
            if (dataset in test_data_ex and (each_csv in ['dataset1.csv','dataset2.csv'])) or (dataset not in test_data_ex and (each_csv in ['dataset1.csv','dataset2.csv','dataset3.csv'])):
                test_instances.append(np.concatenate(test_segs, axis=None))
                test_label.append(dataset)
            elif each_csv[-4:]==".csv":
                train_instances.append(np.concatenate(train_segs, axis=None))
                train_label.append(dataset)
                
    test_df = pd.DataFrame(test_instances, columns=all_features)
    test_df = test_df.loc[:,feature_list]
    test_df["label"] = test_label
    train_df = pd.DataFrame(train_instances, columns=all_features)
    train_df = train_df.loc[:,feature_list]
    train_df["label"] = train_label
    return train_df,test_df
```

#### 1. (d). iii. Break each time series in your training set into l ∈ {1, 2, . . . , 20} time series of approximately equal length and use logistic regression to solve the binary classification problem, using time-domain features.

```python
#function to perform logistic regression with cross validation
def logisticRegressionCrossValidation(l, downsample=False):
    feature_list = ['min','max','mean','median','std','first_quartile', 'third_quartile']
    pass_features = []
    for j in range(len(column_names)*l):
        for i in feature_list:
            pass_features.append(i+'_'+str(j+1))

    #train test split
    train_dataset,test_dataset = split_train_test(pass_features,l)
    
    bending_train = pd.DataFrame(train_dataset)
    bending_train['label']=bending_train['label'].replace('bending1',"1")
    bending_train['label']=bending_train['label'].replace('bending2',"1")
    bending_train.loc[bending_train['label'] != "1", 'label'] = "0"
    
    #downsample majority class
    if downsample:
        class0 = bending_train[bending_train["label"]=="0"]
        class1 = bending_train[bending_train["label"]=="1"]
        majority = ""
        if len(class1)>len(class0):
            majority = "1"
            major_df = class1
            minor_df = class0
        else:
            majority = "0"
            major_df = class0
            minor_df = class1
        major_df = resample(major_df, replace=False, n_samples=len(minor_df), random_state=42)
        bending_train = major_df.append(minor_df, ignore_index=True)        
    
    xtrain = bending_train.iloc[:,:-1]
    ytrain = bending_train.iloc[:,-1]
    
    #stratified k fold with 5 splits
    cross_valid = StratifiedKFold(shuffle=True, random_state=7, n_splits=5)
    
    #penalty set as 'none'
    model = LogisticRegression(max_iter=100, penalty='none')
    selector = RFECV(estimator=model, cv=cross_valid, scoring='accuracy')
    selector.fit(xtrain, ytrain)
    
    return selector, model
```

#### 1. (d). iii. Use 5-fold cross-validation to determine the best value of the pair (l, p), where p is the number of features used in recursive feature elimination.

```python
for i in range(1,21):
    selector, model = logisticRegressionCrossValidation(i)
    print("Pair:(",str(i)+","+str(selector.n_features_)+")", "Scores:", max(selector.grid_scores_))
```

Pair (1,6) is the best value pair since it has the highest score

#### 1. (e). i. Repeat 1(d)iii using L1-penalized logistic regression,8 i.e. instead of using p- values for variable selection, use L1 regularization.

```python
#function to perform L1 Regression, Gaussian Naive Bayes and Multinomial Naive Bayes 
def L1Regression(l, model, plot=False, multiclass=False):
    feature_list = ['min','max','mean','median','std','first_quartile', 'third_quartile']
    pass_features = []
    for j in range(len(column_names)*l):
        for i in feature_list:
            pass_features.append(i+'_'+str(j+1))

    train_dataset,test_dataset = split_train_test(pass_features,l)

    bending_train = pd.DataFrame(train_dataset)
    bending_test = pd.DataFrame(test_dataset)
    
    if not multiclass:
        bending_train['label']=bending_train['label'].replace('bending1',"1")
        bending_train['label']=bending_train['label'].replace('bending2',"1")
        bending_train.loc[bending_train['label'] != "1", 'label'] = "0"   

        bending_test['label']=bending_test['label'].replace('bending1',"1")
        bending_test['label']=bending_test['label'].replace('bending2',"1")
        bending_test.loc[bending_test['label'] != "1", 'label'] = "0"
    elif multiclass:
        bending_train['label']=bending_train['label'].replace('bending1',"bending")
        bending_train['label']=bending_train['label'].replace('bending2',"bending")
        
        bending_test['label']=bending_test['label'].replace('bending1',"bending")
        bending_test['label']=bending_test['label'].replace('bending2',"bending")

    xtrain = bending_train.iloc[:,:-1]
    ytrain = bending_train.iloc[:,-1]
    
    xtest = bending_test.iloc[:,:-1]
    ytest = bending_test.iloc[:,-1]
    
    scaler = MinMaxScaler()
    scaler.fit(xtrain)
    xtrain = scaler.transform(xtrain)
    xtest = scaler.transform(xtest)
    
    cross_valid = StratifiedKFold(shuffle=True, random_state=7, n_splits=5)
    if not multiclass and model=="logistic_regression":
        model = LogisticRegressionCV(max_iter=100, penalty="l1", solver="liblinear", cv=cross_valid)
    elif multiclass and model=="logistic_regression":
        model = LogisticRegressionCV(max_iter=100, penalty="l1", solver="saga", cv=cross_valid)
    elif model=="gaussian_nb":
        model = GaussianNB()
    elif model=="multinomial_nb":
        model = MultinomialNB()
    
    model.fit(xtrain, ytrain)
    
    train_score = model.score(xtrain, ytrain)
    test_score = model.score(xtest,ytest)
    
    if not plot:
        print("l value:",l,"  train set score:",str(train_score)[:5],"  test set score:",str(test_score)[:5])
    elif plot and not multiclass:
        ypred = model.predict(xtrain)

        confusion_mat = confusion_matrix(ytrain,ypred)
        ypred_probab = model.predict_proba(xtrain)

        fpr, tpr, thresholds = roc_curve(ytrain.to_numpy(dtype="int"),ypred_probab[:,1])
        roc_auc = auc(fpr, tpr)

        print("Train Confusion Matrix")
        print(confusion_mat)
        
        plt.figure()
        plt.title('ROC Curve Training Set')
        plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
        plt.legend(loc = 'lower right')
        plt.plot([0, 1], [0, 1],'r--')
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.show()  
        
        ypred = model.predict(xtest)

        confusion_mat = confusion_matrix(ytest,ypred)
        ypred_probab = model.predict_proba(xtest)

        fpr, tpr, thresholds = roc_curve(ytest.to_numpy(dtype="int"),ypred_probab[:,1])
        roc_auc = auc(fpr, tpr)

        print("Test Confusion Matrix")
        print(confusion_mat)
        
        plt.figure()
        plt.title('ROC Curve Test Set')
        plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
        plt.legend(loc = 'lower right')
        plt.plot([0, 1], [0, 1],'r--')
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.show()
    
    elif multiclass and plot:
        ypred = model.predict(xtrain)

        confusion_mat = confusion_matrix(ytrain,ypred)
        ypred_probab = model.predict_proba(xtrain)
        
        plt.figure()
        for i,cl in enumerate(model.classes_):
            fpr, tpr, _ = roc_curve(ytrain, ypred_probab[:, i], pos_label=cl)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=cl+' ROC curve (area = %0.2f)' % roc_auc)


        print("Train Confusion Matrix")
        print(confusion_mat)
        
        plt.title('ROC Curve Training Set')
        plt.legend(loc = 'lower right')
        plt.plot([0, 1], [0, 1],'r--')
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.show() 
        
        ypred = model.predict(xtest)

        confusion_mat = confusion_matrix(ytest,ypred)
        ypred_probab = model.predict_proba(xtest)
        
        plt.figure()
        for i,cl in enumerate(model.classes_):
            fpr, tpr, _ = roc_curve(ytest, ypred_probab[:, i], pos_label=cl)
            roc_auc = auc(fpr, tpr)
            plt.plot(fpr, tpr, label=cl+' ROC curve (area = %0.2f)' % roc_auc)


        print("Test Confusion Matrix")
        print(confusion_mat)
        
        plt.title('ROC Curve Test Set')
        plt.legend(loc = 'lower right')
        plt.plot([0, 1], [0, 1],'r--')
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.show()
    
```

#### 2. ISLR 3.7.4

I collect a set of data (n = 100 observations) containing a single predictor and a quantitative response. I then fit a linear regression model to the data, as well as a separate cubic regression, i.e. Y = β0 +β1X +β2X2 +β3X3 +ε.


(a) Suppose that the true relationship between X and Y is linear, i.e. Y = β0 + β1X + ε. Consider the training residual sum of squares (RSS) for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.

The cubic regression model will have more features than linear regression providing a better fit to the training data. Thus, cubic regression training RSS will be lower.

(b) Answer (a) using test rather than training RSS.

The cubic regression model will have more features so it will overfit the training data compared to linear regression. So, linear regression test RSS will be lower since the relationship between X and Y is linear.

(c) Suppose that the true relationship between X and Y is not linear, but we don’t know how far it is from linear. Consider the training RSS for the linear regression, and also the training RSS for the cubic regression. Would we expect one to be lower than the other, would we expect them to be the same, or is there not enough information to tell? Justify your answer.

Since X and Y are non-linear, cubic regression will have less training RSS.

(d) Answer (c) using test rather than training RSS.

If the relationship is more cubic than linear, then test RSS for cubic regression will be lower. If the relationship is more linear than cubic, then test RSS for linear regression will be lower.
