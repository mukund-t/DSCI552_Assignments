#### 1. (i) Fit a PCR model on the training set, with M (the number of principal compo- nents) chosen by cross-validation. Report the test error obtained.

https://scikit-learn.org/stable/auto_examples/cross_decomposition/plot_pcr_vs_pls.html

```python
ncomp_test_errors=[]
for n_comp in range(1,123):
    kf = KFold(5,shuffle=True,random_state=7)
    test_errors = []
    for train_index, test_index in kf.split(xtrain):
        cv_xtrain, cv_xtest = xtrain[train_index], xtrain[test_index]
        cv_ytrain, cv_ytest = ytrain[train_index], ytrain[test_index]

        pcr = make_pipeline(StandardScaler(), PCA(n_components=n_comp), LinearRegression())
        pcr.fit(cv_xtrain, cv_ytrain)

        cv_ytest_pred = pcr.predict(cv_xtest)
        cv_test_error = mean_squared_error(cv_ytest_pred, cv_ytest)
        test_errors.append(cv_test_error)
    ncomp_test_errors.append(np.mean(test_errors))
print("M:",ncomp_test_errors.index(min(ncomp_test_errors))+1)
print("CV error at M:",min(ncomp_test_errors))
```

```python
pcr = make_pipeline(StandardScaler(), PCA(n_components=95), LinearRegression())
pcr.fit(xtrain,ytrain)

ytrain_pred = pcr.predict(xtrain)
train_error = mean_squared_error(ytrain,ytrain_pred)

ytest_pred = pcr.predict(xtest)
test_error = mean_squared_error(ytest,ytest_pred)

print("Train error:", train_error)
print("Test error:", test_error)
```

#### 2. (c) Train a random forest to classify the data set. Do NOT compensate for class imbalance in the data set. Calculate the confusion matrix, ROC, AUC, and misclassification for training and test sets and report them. Calculate Out of Bag error estimate for your random forset and compare it to the test error.

```python
ypred = random_forest.predict(xtrain)

confusion_mat = confusion_matrix(ytrain,ypred)
ypred_probab = random_forest.predict_proba(xtrain)

fpr, tpr, thresholds = roc_curve(ytrain,ypred_probab[:,1])
roc_auc = auc(fpr, tpr)

print("Confusion Matrix")
print(confusion_mat)

plt.title('ROC Curve Training Set')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.4f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
ypred = random_forest.predict(xtest)

confusion_mat = confusion_matrix(ytest,ypred)
ypred_probab = random_forest.predict_proba(xtest)

fpr, tpr, thresholds = roc_curve(ytest,ypred_probab[:,1])
roc_auc = auc(fpr, tpr)

print("Confusion Matrix")
print(confusion_mat)

plt.title('ROC Curve Test Set')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.4f' % roc_auc)
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
from sklearn.metrics import accuracy_score

print("OOB error for training set:",round((1-random_forest.oob_score_),4))
print("Test error:",round(1-accuracy_score(ytest,ypred),4))
```

```python
from weka.filters import Filter
smote = Filter(classname="weka.filters.supervised.instance.SMOTE")

cls = Classifier(classname="weka.classifiers.trees.LMT")

from weka.classifiers import FilteredClassifier

fc = FilteredClassifier()
fc.filter = smote
fc.classifier = cls

evl = Evaluation(weka_train)
evl.crossvalidate_model(fc, weka_train, 5, Random(1))

```

```python
print("Score",evl.percent_correct)
```

```python
print(evl.summary())
print(evl.class_details())
```

```python
print("Training confusion matrix")
print(evl.confusion_matrix)
```

```python
import weka.plot.classifiers as plcls
plcls.plot_roc(evl, class_index=[0, 1], wait=True)
```

```python
fc.build_classifier(weka_train)
```

```python
evaluation = Evaluation(weka_test)
test_evl = evaluation.test_model(fc, weka_test)
print(test_evl)
print(evaluation.summary())
```

```python
print("Test confusion matrix")
print(evaluation.confusion_matrix)
```

```python
print(evaluation.class_details())
```

```python
plcls.plot_roc(evaluation, class_index=[0, 1], wait=True)
```

For this dataset, when compared to SMOTE the uncompensated case performs better since the AUC in the ROC curve for test set is less for SMOTE. By observing the confusion matrix, we can notice that SMOTE has reduced the misclassification rate for the positive class.

#### 3. ISLR 6.6.3

a. iv. Steadily decrease. If we increase s from 0, once it reaches a large value the beta that satisfies and minimizes the RSS will always be the least squares solution. So, until that s value is reached, the train RSS will steadily decrease.

b. ii. Decrease initially, and then eventually start increasing in a U shape. When s increases the flexibility of model increases, so the test RSS decreases until a point and then starts to become more overfit.

c. iii. Steadily increase. When s increases, it approaches the least squares solution which has higher variance. As the model flexibility increases, variance also increases.

d. iv. Steadily decrease. When s increases, the flexibility of the model increases which leads to gradual decrease in bias.

e. v. Remain constant. The irreducible error is expected for any model as the error is introduced by noise for any model. It will remain constant regardless of model flexibility.
