#### 1. (h) Improve your model using possible interaction terms or nonlinear asso- ciations between the predictors and response

#### Linear Regression model with all possible interactions and non-linearities

```python
interaction_df = df
mse_dict ={}

idx=5
mapping_dict={}
for i in column_names[:-1]:
    for j in column_names[:-1]:
        if i==j:
            interaction_df[i+" x "+j]=df[i].to_numpy()*df[j].to_numpy()
            mapping_dict["x"+str(idx)]=i+"^2"
            idx+=1
        if i!=j and {i,j} not in mapping_dict.values():
            interaction_df[i+" x "+j]=df[i].to_numpy()*df[j].to_numpy()
            mapping_dict["x"+str(idx)]={i,j}
            idx+=1

pe = interaction_df.pop("PE")
interaction_df["PE"]=pe

interaction_df
```

```python
train_interaction, test_interaction = train_test_split(interaction_df, test_size=0.3,random_state=42)

train_interaction_x = train_interaction.iloc[:,:-1].values
train_interaction_y = train_interaction.iloc[:,-1].values
test_interaction_x = test_interaction.iloc[:,:-1].values
test_interaction_y = test_interaction.iloc[:,-1].values


train_interaction_xconst = sm.add_constant(train_interaction_x)
test_interaction_xconst = sm.add_constant(test_interaction_x)
model = sm.OLS(train_interaction_y, train_interaction_xconst)
res = model.fit()

train_interaction_y_pred = res.predict(train_interaction_xconst)
test_interaction_y_pred = res.predict(test_interaction_xconst)

mse_dict['All interactions and non-linearities linear regression'] = {
    "Train MSE":mean_squared_error(train_interaction_y,train_interaction_y_pred),
    "Test MSE":mean_squared_error(test_interaction_y,test_interaction_y_pred)
}

print("Train MSE:",mean_squared_error(train_interaction_y,train_interaction_y_pred))
print("Test MSE:",mean_squared_error(test_interaction_y,test_interaction_y_pred))

res.summary()
```

The features APxAT, V^2, APxV, RHxV are insignificant if we consider p > |t| less than 0.01 as insignificant. AT, V needs to be retained due to hierarchy principle even though they have large p value.

```python
mapping_dict
```

#### Simple Linear Regression model

```python
linear_train_x = train_interaction[['AT','V','AP','RH']].to_numpy()
linear_train_y = train_interaction[['PE']].to_numpy()
linear_test_x = test_interaction[['AT','V','AP','RH']].to_numpy()
linear_test_y = test_interaction[['PE']].to_numpy()

linear_train_xconst = sm.add_constant(linear_train_x)
linear_test_xconst = sm.add_constant(linear_test_x)
model = sm.OLS(linear_train_y, linear_train_xconst)
res = model.fit()

linear_train_y_pred = res.predict(linear_train_xconst)
linear_test_y_pred = res.predict(linear_test_xconst)

mse_dict['Simple linear regression'] = {
    "Train MSE":mean_squared_error(linear_train_y,linear_train_y_pred),
    "Test MSE":mean_squared_error(linear_test_y,linear_test_y_pred)
}

print("Train MSE:",mean_squared_error(linear_train_y,linear_train_y_pred))
print("Test MSE:",mean_squared_error(linear_test_y,linear_test_y_pred))

res.summary()
```

#### Linear Regression model with all possible interactions and non-linearities after removing insignificant terms

```python
train_interaction = train_interaction.drop(['AT x AP', 'V x V', 'V x AP', 'V x RH'], axis=1)
test_interaction = test_interaction.drop(['AT x AP', 'V x V', 'V x AP', 'V x RH'], axis=1)

train_interaction_x = train_interaction.iloc[:,:-1].values
train_interaction_y = train_interaction.iloc[:,-1].values
test_interaction_x = test_interaction.iloc[:,:-1].values
test_interaction_y = test_interaction.iloc[:,-1].values


train_interaction_xconst = sm.add_constant(train_interaction_x)
test_interaction_xconst = sm.add_constant(test_interaction_x)
model = sm.OLS(train_interaction_y, train_interaction_xconst)
res = model.fit()

train_interaction_y_pred = res.predict(train_interaction_xconst)
test_interaction_y_pred = res.predict(test_interaction_xconst)

mse_dict['All interactions and non-linearities linear regression removed insignificant features'] = {
    "Train MSE":mean_squared_error(train_interaction_y,train_interaction_y_pred),
    "Test MSE":mean_squared_error(test_interaction_y,test_interaction_y_pred)
}

print("Train MSE:",mean_squared_error(train_interaction_y,train_interaction_y_pred))
print("Test MSE:",mean_squared_error(test_interaction_y,test_interaction_y_pred))

res.summary()
```

#### 1. (i) KNN Regression

#### KNN Regression with raw features

```python
#Raw Features
k_list = []
kinv_list=[]
train_error_list=[]
test_error_list=[]
for k in range(1,101):
    neigh = KNeighborsRegressor(n_neighbors=k)
    neigh.fit(linear_train_x,linear_train_y)
    linear_train_ypred = neigh.predict(linear_train_x)
    linear_test_ypred = neigh.predict(linear_test_x)
    
    train_error = mean_squared_error(linear_train_y,linear_train_ypred)
    test_error = mean_squared_error(linear_test_y,linear_test_ypred)
    
    train_error_list.append(train_error)
    test_error_list.append(test_error)
    k_list.append(k)
    kinv_list.append(1/k)

opt_k = test_error_list.index(min(test_error_list))+1
print("Optimal K:",opt_k)
print("Train Error:",train_error_list[opt_k-1])
print("Test Error:",test_error_list[opt_k-1])

mse_dict['KNN raw features'] = {
    "Train MSE":train_error_list[opt_k-1],
    "Test MSE":test_error_list[opt_k-1]
}
```

```python
plt.plot(kinv_list,train_error_list,label="Train Error")
plt.plot(kinv_list,test_error_list,label="Test Error")
plt.xlabel("1/K")
plt.ylabel("Error")
plt.title("KNN with raw features")
plt.legend()
```

#### KNN Regression with scaled features

```python
#Scaled Features

scaler = MinMaxScaler()
scaler.fit(linear_train_x)
linear_train_xsc = scaler.transform(linear_train_x)
linear_test_xsc = scaler.transform(linear_test_x)

k_list = []
kinv_list=[]
train_error_list=[]
test_error_list=[]
for k in range(1,101):
    neigh = KNeighborsRegressor(n_neighbors=k)
    neigh.fit(linear_train_xsc,linear_train_y)
    linear_train_ypred = neigh.predict(linear_train_xsc)
    linear_test_ypred = neigh.predict(linear_test_xsc)
    
    train_error = mean_squared_error(linear_train_y,linear_train_ypred)
    test_error = mean_squared_error(linear_test_y,linear_test_ypred)
    
    train_error_list.append(train_error)
    test_error_list.append(test_error)
    k_list.append(k)
    kinv_list.append(1/k)

opt_k = test_error_list.index(min(test_error_list))+1
print("Optimal K:",opt_k)
print("Train Error:",train_error_list[opt_k-1])
print("Test Error:",test_error_list[opt_k-1])

mse_dict['KNN scaled features'] = {
    "Train MSE":train_error_list[opt_k-1],
    "Test MSE":test_error_list[opt_k-1]
}
```

```python
plt.plot(kinv_list,train_error_list,label="Train Error")
plt.plot(kinv_list,test_error_list,label="Test Error")
plt.xlabel("1/K")
plt.ylabel("Error")
plt.title("KNN with scaled features")
plt.legend()
```

### 2. 2.4.1

#### 1.a. The sample size n is extremely large, and the number of predictors p is small.
A flexible statistical model is better since when the sample size is large, an inflexible model will underfit the data.
#### 1.b. The number of predictors p is extremely large, and the number of observations n is small.
A flexible model will overfit the data, so an inflexible model is better.
#### 1.c. The relationship between the predictors and response is highly non-linear.
A flexible model is better because flexible models result in non-linear fit to the data.
#### 1.d. The variance of the error terms, i.e. Ïƒ2 = Var(Îµ), is extremely high.
A flexible model will overfit as it tries to fit the variance of error terms, so an inflexible model is better.

```python
data = [[0,3,0,'Red'],
        [2,0,0,'Red'],
        [0,1,3,'Red'],
        [0,1,2,'Green'],
        [-1,0,1,'Green'],
        [1,1,1,'Red']]
test_p = [0,0,0]
dist = []
idx = 0
for each_data in data:
    temp = np.linalg.norm(np.array(test_p)-np.array(each_data[:-1]))
    dist.append(temp)
    print("Distance to point from data "+str(idx)+":",temp)
    idx+=1
```
